```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Anaconda JYP\backpacks.csv")

# 1. Basic info
print(df.shape)        # rows, columns
print(df.columns)      # column names
print(df.dtypes)       # data types

# 2. Missing values check
print(df.isnull().sum())

# 3. Remove duplicates
df.drop_duplicates(inplace=True)

# 4. Clean price column (remove ₹ and quotes)
df['final_price'] = df['final_price'].str.replace("₹","").str.replace(",","").str.replace('"','').astype(float)

# 5. Derived column: discounted price check
df['calculated_final_price'] = df['initial_price'] - (df['initial_price'] * df['discount']/100)

# 6. Quick stats
print(df[['rating','final_price','discount']].describe())

# 7. Save cleaned dataset
df.to_csv(r"C:\Anaconda JYP\backpacks_cleaned.csv", index=False)

# 8. Preview cleaned data
print(df.head())
```

```python
## Brief Summary
In this assignment, I loaded the shopping dataset into a Pandas DataFrame and explored its structure using 
head, shape, and data types. I identified and handled missing values, removed duplicates, and cleaned the
price column by converting it into numeric format. I also created a derived column(calculated_final_price)
to check discounted prices. Finally, I saved the cleaned dataset as a new CSV file. This process helped me
understand basic data exploration and cleaning steps in Pandas.
```
